```
import pandas as pd
import numpy as np
from scipy.io import arff

# 1. Cargar el archivo ARFF
# Asegúrate de haber extraído el archivo 'chronic_kidney_disease_full.arff' en tu carpeta de trabajo
data, meta = arff.loadarff('chronic_kidney_disease_full.arff')
df = pd.DataFrame(data)

# 2. Limpieza de strings (los archivos ARFF cargan los textos como bytes b'text')
for col in df.columns:
    if df[col].dtype == object:
        df[col] = df[col].str.decode('utf-8').str.strip().str.lower()

# 3. Reemplazar signos de interrogación '?' (comunes en ARFF) por NaN (valores faltantes)
df.replace('?', np.nan, inplace=True)

# 4. Convertir columnas numéricas que se hayan detectado erróneamente como texto
# Nota: Revisa tu archivo .info.txt para mapear cuáles son numéricas (ej. age, bp, bgr)
num_cols = ['age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc']
for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# --- REPORTE INICIAL DE CALIDAD DE DATOS ---
print("--- ESTRUCTURA DEL DATASET ---")
print(f"Registros (Filas): {df.shape[0]} | Variables (Columnas): {df.shape[1]}\n")

print("--- TIPOS DE VARIABLES Y VALORES FALTANTES ---")
calidad = pd.DataFrame({
    'Tipo': df.dtypes,
    'Valores Faltantes (N)': df.isnull().sum(),
    'Valores Faltantes (%)': (df.isnull().sum() / len(df)) * 100
})
print(calidad)

print("\n--- REGISTROS DUPLICADOS ---")
print(f"Cantidad de filas duplicadas: {df.duplicated().sum()}")
```
